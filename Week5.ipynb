#### Question: Write a short Python program that displays the names of all of the movies from 1959. You’ll get full credit at this point for displaying the full lines that contain this information. “Bonus style points” if you display just the movie names.

##### Answer: According to teh README file, the file called 'u.item' contained all the movie title information.         Furthermore, after reviewing the contents of this file, we're interested in the 2nd and third column:.  Movie Title and Release Date.         Lastly, this is a "pipe" or "|" deliminated text file.

```python
# Basic logic to this program:
# 1. Open the file
# 2. Loop through every line in the file and do the following:
#   2.1. Use the "Split" method to divy up the line into a list variable so you can index columns
#   2.2. Get the year out of the 2nd index (i.e. 3rd column since indexes start at 0)
#   2.3. Print the 1st index (2nd column) a.k.a. Movie Title if the Year >= 1959

fhand = open('C:/Users/ssadalge/u.item')

for line in fhand:
    line = line.rstrip()    # in hindsight, this line really not needed but got into the habit of this from practice problems in the book
    columns = line.split("|")
    year = columns[2][7:]   # I checked and all the dates are 11 chars: DD-MMM-YYYY.  If they weren't I could use columns[2][len(columns[2])-4:] to get the right 4 positions)
    if year >= '1959':
        print columns[1]
        
```
